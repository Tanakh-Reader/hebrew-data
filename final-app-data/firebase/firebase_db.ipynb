```python
import firebase_admin
from firebase_admin import credentials, firestore
import csv
```

```python
f = 'firebase-credentials.json'
cred = credentials.Certificate(f)
firebase_admin.initialize_app(cred)

db = firestore.client()
```

```python
def get_books() -> list[dict]:

    books:list[dict] = []

    with open('/Users/sethhowell/Desktop/Hebrew-Literacy-App/hebrew-data/data-staging/data/data_files/books.csv', 'r', encoding='utf-8-sig') as csv_file:
        
        return list( csv.DictReader(csv_file, delimiter=',') )
```

```python
def get_chapter(bk, ch):
    
    data = db.collection("bible").document(str(bk)) \
            .collection('chapters').document(str(ch)) \
            .get()
        
    return(data)
```
